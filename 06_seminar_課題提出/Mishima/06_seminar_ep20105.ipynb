## 課題3

```python
import matplotlib.pyplot as plt
import numpy as np

#シグモイド関数
def sigmoid(x):
  return 1/(1+np.exp(-x))

#ソフトマックス関数
def softmax(x):
  exp_x=np.exp(x)
  sum_exp_x=np.sum(exp_x)
  return exp_x/sum_exp_x

#tanh関数
def tanh(x):
  exp_x=np.exp(x)
  exp_subx=np.exp(-x)
  return (exp_x-exp_subx)/(exp_x+exp_subx) 

x=np.arange(-5.0,5.0,0.1)
sigmoid=sigmoid(x)
softmax=softmax(x)
tanh=tanh(x)

plt.plot(x,sigmoid,label="sigmoid",color='red')  #sigmoid関数表示
plt.plot(x,softmax,label="softmax",color='blue') #softmax関数表示
plt.plot(x,tanh,label="tanh",color='green')      #tanh関数表示

plt.grid()   #グリッド表示
plt.legend() #凡例表示

plt.ylim(-1.2,1.2)
plt.title("activation function")
plt.xlabel("input")
plt.ylabel("output")
```
