```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-5.0,5.0,0.1)

def sigmoid(x):
  return 1/(1+np.exp(-x))

def softmax(x):
  exp_x = np.exp(x)
  sum_exp_x = np.sum(exp_x)
  return exp_x/ sum_exp_x

def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

plt.plot(x,sigmoid(x),label="sigmoid",color='blue')
plt.plot(x,softmax(x),label="softmax",color='green')
plt.plot(x,tanh(x),label="tanh",color='orange')

plt.ylim(-1.2,1.2)
plt.title("sigmoid,softmax,tanh")
plt.xlabel("imput")
plt.ylabel("output")
plt.grid()
plt.legend()
plt.show()
```
