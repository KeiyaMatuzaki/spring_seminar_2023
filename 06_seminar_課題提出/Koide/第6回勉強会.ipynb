**課題3**

シグモイド関数・ソフトマックス関数・tanh関数をまとめて表示してください

条件

x = np.arange(-5.0,5.0,0.1),plt.ylim(-1.2,1.2),凡例，軸ラベル，タイトル

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5.0,5.0,0.1)

def sigmoid(x):
  return 1/(1+np.exp(-x))

def softmax(x):
  return np.exp(x) / np.sum(np.exp(x))

def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))


sigmoid = sigmoid(x)
softmax = softmax(x)
tanh = tanh(x)
plt.plot(x,sigmoid,label="sigmoid",color='red')
plt.plot(x,softmax,label="softmax",color='blue')
plt.plot(x,tanh,label="tanh",color='green')

plt.ylim(-1.2,1.2)

plt.title("function")
plt.xlabel("imput")
plt.ylabel("output")
plt.grid()
plt.legend()
plt.show()


```
